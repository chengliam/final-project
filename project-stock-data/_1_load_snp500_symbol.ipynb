```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy import Table, Column, MetaData, Integer, String, Date, Float, BigInteger

url = 'https://raw.githubusercontent.com/datasets/s-and-p-500-companies/master/data/constituents.csv'
df = pd.read_csv(url)
print(df.head)

engine = create_engine('postgresql+psycopg2://postgres:admin1234@localhost:5432/bootcamp_2504p')

metadata = MetaData()

stock_data = Table(
  'stocks',
  metadata,
  Column("Symbol", String(10), primary_key=True)
)

stocks_df = df[['Symbol']].rename(columns={'Symbol' : 'symbol'})
stocks_df.to_sql('stocks', con=engine, if_exists='replace', index=False)

print("S&P 500 symbols loaded successfully!")

stock_profiles_data = Table(
  'stock_profiles',
  metadata,
  Column("Symbol", String(10), primary_key=True),
  Column("Security", String(20), nullable=False),
  Column("GICS Sector", String(20), nullable=False),
  Column("GICS Sub-Industry", String(20), nullable=False),
  Column("Headquarters Location", String(20), nullable=False),
  Column("Date added", Date, nullable=False),
  Column("CIK", Integer, nullable=False)
)

stock_profiles_df = df[['Symbol', 'Security', 'GICS Sector', 'GICS Sub-Industry', 'Headquarters Location', 'Date added', 'CIK', 'Founded']].rename(columns={'Symbol': 'symbol', 'Security': 'security', 'GICS Sector': 'gics_sector', 'GICS Sub-Industry': 'gics_sub_industry', 'Headquarters Location': 'hq_location', 'Date added': 'date_added', 'CIK': 'cik', 'Founded': 'founded'})
stock_profiles_df.to_sql('stock_profiles', con=engine, if_exists='replace', index=False)

print("S&P 500 symbols loaded successfully!")
```
